### 직선
직선의 종류 (line type)

1. cv2.LINE_4 : 상하좌우 4방향으로 연결된 선
2. cv2.LINE_8 : 대각선을 포함한 8방향으로 연결된 선(기본값)
3. cv2.LINE_AA : 부드러운 선(anti-aliasing)

#라인 4와 라인 8의 차이 상ㅎ
라인 4는 상하좌우로 선을 이으기 때문에 다 연결이 되어있음
라인 8은 그림판처럼 좀 굳이 연결이 안디있어도 도밈

```python
import cv2
import numpy as np

img = np.zeros((480,640,3),dtype = np.uint8)
```

```python
COLOR = (0,255,255) #yellow
THICKNESS =3

cv2.line(img, (50,100),(400,50), COLOR, THICKNESS, cv2.LINE_8) #50,100부터 400,50까지
#그릴 위치, 시작점 끝점 색깔 두께 선 종류
cv2.line(img, (50,100),(400,50), COLOR, THICKNESS, cv2.LINE_8) #50,100부터 400,50까지
cv2.line(img, (50,200),(400,150), COLOR, THICKNESS, cv2.LINE_8) #50,100부터 400,50까지
cv2.line(img, (50,300),(400,250), COLOR, THICKNESS, cv2.LINE_8) #50,100부터 400,50까지
 
cv2.imshow('img',img) #여기서 img를 안해서 계속 오류가 떴었네..
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 원

```python
img = np.zeros((480,640,3), dtype = np.uint8) #0은 흰색
color = (255,255,0) #BGR 임 ㅎㅎㅎ 초록색과 파란색을 섞은 거임
RADIUS = 50 # 반지름
THICKNESS = 10 #두께

cv2.circle(img, (200,100),RADIUS,color,THICKNESS, cv2.LINE_AA)
cv2.circle(img, (400,200),RADIUS,color,cv2.FILLED, cv2.LINE_AA) #속이 꽉찬 원을 만들고 싶어 그러면 이럴때 cv2.FIllED대신 -1해 돋됨

#세로기준 #가로기준 
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 사각형

```python
img = np.zeros((480,640,3), dtype = np.uint8) #0은 흰색
color = (0,255,0) #BGR 임 ㅎㅎㅎ 초록색과 파란색을 섞은 거임

THICKNESS = 3 #두께

cv2.rectangle(img,(100,100),(200,200),color,THICKNESS) # 속이 빈 사각형임
cv2.rectangle(img,(200,200),(300,300),color,cv2.FILLED) # 속이 빈 사각형임

#세로기준 #가로기준 
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 다각형

```python
img = np.zeros((480,640,3), dtype = np.uint8) #0은 흰색

color = (0,0,255) #BGR 임 빨간색은 
THICKNESS = 3 #두께

pts1 = np.array([[100,100],[200,100],[100,200]])
pts2 = np.array([[200,100],[300,100],[300,200]])
cv2.polylines(img, [pts1],False, color, THICKNESS, cv2.LINE_AA) 
#pts1을 리스트로 감싸주는 게 포인트!!, False하면 첫 점과 끝점을 안이어줌, 
cv2.polylines(img, [pts2],True, color, THICKNESS, cv2.LINE_AA) 
#세로기준 #가로기준 
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.polylines(img, [pts1, pts2],True, color, THICKNESS, cv2.LINE_AA) 
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pts3 = np.array([[[100,300],[200,300],[100,400]],[[200,300],[300,300],[300,400]]])
cv2.fillPoly(img, pts3, color, cv2.LINE_AA)
#그릴 위치, 그릴좌표, 색깔, 선 종류
# -- 두께도 필요업음 어차피 그릴거니까
# -- 닫힘 여부도 필요 없음
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
