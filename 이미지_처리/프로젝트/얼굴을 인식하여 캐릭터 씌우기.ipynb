## Face Detectiton vs Face Recognition

Face Detection vs Face Recognition 출결을 할려면 인지까지 해야곘죠!!!

```python
import cv2
import mediapipe as mp

# 얼굴을 찾고, 찾은 얼굴에 표시를 해주기 위한 변수 정의
mp_face_detection = mp.solutions.face_detection #얼굴 검출을 위한 face_detection 모듈을 사용
mp_drawing = mp.solutions.drawing_utils # 얼굴의 특징을 그리기 위한 draawing_utilis 모듈 사용

# 동영상 파일 열기
cap = cv2.VideoCapture('face_video.mp4')
# fps = cap.get(cv2.CAP_PROP_FPS)*0.5

#
with mp_face_detection.FaceDetection(
    model_selection =0, min_detection_confidence=0.5) as face_detection:
    while cap.isOpened():
        success, image = cap.read()
        if not success:
            print('Ignoring empty camera frame')
            break
        #cam이면 continue도 상관없음


        image.flags.writeable=False
        image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        result = face_detection.process(image)
        if result.detections:
            for detection in result.detections:
                mp_drawing.draw_detection(image, detection)
                
        cv2.imshow('MedaiPipe Face Detection', cv2.flip(image,1)) #cv2.resize(image,None, fx=0.5,fy=0.5)
        if cv2.waitKey(30) ==ord('q'):
            break
            
cap.release()
cv2.destroyAllWindows()
```
